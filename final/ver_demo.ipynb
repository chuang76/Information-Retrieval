```
model = FastText()
model.load_model('/content/drive/My Drive/cc.zh.300.bin')
```

```
with io.open('/content/drive/My Drive/1_Question.json', mode='r', encoding='utf-8') as f:    # 正式 demo 的時候記得改檔名
	JsonList = json.load(f)

def sim(v1, v2):
  cosine = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
  return cosine
```

```
ans = []
ans_cosine = []

stopword_set = set()
with io.open('/content/drive/My Drive/stopword.txt','r', encoding='utf-8') as stopwords:
  for stopword in stopwords:
    stopword_set.add(stopword.strip('\n'))

for i in range(200):

  arr = []
  s = jieba.cut(JsonList[i]['Question'], cut_all=False)
  for k in s:
    if k not in stopword_set:
      arr.append(k)
  arr = ' '.join(arr)

  sv = model.get_numpy_sentence_vector(arr)
  
  a = JsonList[i]['A']
  av = model.get_numpy_vector(a)
  
  b = JsonList[i]['B']
  bv = model.get_numpy_vector(b)
  
  c = JsonList[i]['C']
  cv = model.get_numpy_vector(c)
  
  ascore = sim(sv, av)
  bscore = sim(sv, bv)
  cscore = sim(sv, cv)
  ans_cosine.append((ascore, bscore, cscore))
  
  if(ascore > bscore and ascore > cscore):
    print i, 'A'
    ans.append("A")
  elif(bscore > ascore and bscore > cscore):
    print i, 'B'
    ans.append("B")
  else:
    print i, 'C'
    ans.append("C")
```

```
with open("/content/drive/My Drive/exam_answer.json", "w") as ans_file:
    json.dump(ans, ans_file)
```
